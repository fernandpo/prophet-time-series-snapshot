```python
def stationarity_check(ts):
    # Determing rolling statistics
    roll_mean = ts.rolling(12).mean()
    # Plot rolling statistics:
    plt.plot(ts, color='green',label='Original')
    plt.plot(roll_mean, color='blue', label='Rolling Mean')
    plt.legend(loc='best')
    plt.title('Rolling Mean')
    plt.show(block=False)
    # Perform Augmented Dickey-Fuller test:
    print('Augmented Dickey-Fuller test:')
    df_test = adfuller(ts)
    df_output = pd.Series(df_test[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    print("df_output: \n",df_output)
    for key,value in df_test[4].items():
        df_output['Critical Value (%s)'%key] = value
    print(df_output)
```

```python
stationarity_check(df.y)
```

```python
plot_acf(df.y, lags=28)
plot_pacf(df.y, lags=28)
plt.xlabel('lags')
plt.show()
```

```python
p = q = range(1, 4)
print (p)
```

```python
pq = itertools.product(p, q)
for param in pq:
    try:
        mod = ARMA(df.y,order=param)
        results = mod.fit()
        print('ARMA{} - AIC:{}'.format(param, results.aic))
    except:
        continue
```

```python
p = q = range(0, 4)
r = range(0,2)
pqr = itertools.product(p, r, q)
for param in pqr:
    try:
        mod = ARIMA(df.y,order=param)
        results = mod.fit()
        print('ARIMA{} - AIC:{} - MSE:{}'.format(param, results.aic, np.sum(np.square(results.resid))))
    except:
        continue
```

```python
model = sm.tsa.arima.ARIMA(df.y, order=(1,1,1))  
results_MA = model.fit()  
plt.plot(df.y)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('Fitting data _ MSE: %.2f'% (((results_MA.fittedvalues-df.y)**2).mean()))
plt.show()
```

```python
predictions = results_MA.predict('2009-03-31', '2023-12-31')
```

```python
predictions.head()
print(predictions)
```

```python
mod = sm.tsa.statespace.SARIMAX(df.y,
                                order = (2, 0, 3),
                                seasonal_order= (0,0,0,12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()
```

```python
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
residuals = results.resid
```

```python
sm.stats.acorr_ljungbox(residuals, lags=[20], return_df=True)
```
